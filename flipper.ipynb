The flipper puzzle can also be solved by combining elemental solutions.  An elemental solution is a solution that flips exactly one cell.  The solution to the puzzle is the sum of the elemental solutions that flip the cells that are in the down position, modulo 2.

```python
elemental_states = [
    np.array( [[-1,  1,  1],
               [ 1,  1,  1],
               [ 1,  1,  1]]),
    np.array( [[ 1, -1,  1],
               [ 1,  1,  1],
               [ 1,  1,  1]]),
    np.array( [[ 1,  1,  1],
               [ 1, -1,  1],
               [ 1,  1,  1]])
]
```

```python
_ = solve_and_demo( elemental_states[0])
```

```python
_ = solve_and_demo( elemental_states[1])
```

```python
_ = solve_and_demo( elemental_states[2])
```
